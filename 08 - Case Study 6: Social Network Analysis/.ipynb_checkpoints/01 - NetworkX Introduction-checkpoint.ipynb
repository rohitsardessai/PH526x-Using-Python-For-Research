```python
import networkx as nx
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
G = nx.Graph()    #Create instance of an undirected graph

G.add_node(1)
G.add_nodes_from([2, 3])
G.add_nodes_from(["u", "v"])
```

```python
G.nodes()
```

```python
G.add_edge(1, 2)
G.add_edge("u", "v")
G.add_edges_from([(1,3), (1,4), (1,5), (1,6)])
G.add_edge("u", "w")
```

```python
G.nodes
```

```python
G.remove_node(2)
G.nodes()
```

```python
G.remove_nodes_from([4, 4])
G.nodes
```

```python
G.remove_edge(1,3)
G.edges()
```

```python
G.remove_edges_from([(1, 2), ("u", "v")])
G.edges()
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```
