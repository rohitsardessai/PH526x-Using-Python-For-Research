### Erdős-Rényi (ER) Graph

Erdős-Rényi (ER) is a type of random graph model. <br><br>
This family of random graphs has two parameters: N and p. <br>
N - Number of nodes in the graph. <br>
p - Probability that any pair of nodes will be connected.<br>

#### We will write our own ER function.
<br>
The task is to implement an Erdős-Rényi model as a python function.

```python
from scipy.stats import bernoulli
import networkx as nx
import matplotlib.pyplot as plt
```

```python
bernoulli.rvs(0.2)    #Bernoulli dist with prob of success
```

```python
N = 20
p = 0.2
#create an empty graph
G = nx.Graph()
#add all nodes in the graph
G.add_nodes_from(range(N))
#loop over all pairs of nodes
for node1 in G.nodes():
    for node2 in G.nodes():
        if node1 < node2 and bernoulli.rvs(p=p):    #To ensure that each node pair is checked only once.
            G.add_edge(node1, node2)
```

```python
G.number_of_nodes()
```

```python
nx.draw(G, with_labels = 'True', node_color = 'orange', edge = 'grey')
plt.show()
```

```python
def er_graph(N,p):
    """ Generate an ER graph """
    
    G = nx.Graph()
   
    G.add_nodes_from(range(N))
    #loop over all pairs of nodes
    for node1 in G.nodes():
        for node2 in G.nodes():
            if node1 < node2 and bernoulli.rvs(p=p):    #To ensure that each node pair is checked only once.
                G.add_edge(node1, node2)
    return G
```

```python
nx.draw(er_graph(50, 0.08), node_size = 50, node_color = 'gray')
plt.show()
```
