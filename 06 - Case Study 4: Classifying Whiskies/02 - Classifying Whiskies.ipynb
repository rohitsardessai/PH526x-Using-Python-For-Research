```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import SpectralCoclustering
```

```python
whisky = pd.read_csv('whiskies.txt')
whisky['Region'] = pd.read_csv('regions.txt') #add regions as a column
whisky.head()
whisky.iloc[5:10, 0:5]
whisky.columns
```

```python
flavors = whisky.iloc[:,2:14]
flavors
```

```python
corr_flavors = pd.DataFrame.corr(flavors) #Find corelation between the columns
corr_flavors
```

```python
corr_flavors = pd.DataFrame.corr(flavors) #corelation between flavors
plt.figure(figsize = (10, 10))
plt.pcolor(corr_flavors)
plt.colorbar()
plt.show()
plt.savefig("corr_flavors.pdf")
```

```python
flavors.transpose()
corr_whisky = pd.DataFrame.corr(flavors.transpose()) #Find corelation between whiskys
corr_whisky
```

```python
corr_whisky = pd.DataFrame.corr(flavors.transpose())  #Find corelation between whiskys
plt.figure(figsize = (10, 10))
plt.pcolor(corr_whisky)
plt.colorbar()
#plt.axis("tight")
plt.show()
plt.savefig("corr_whisky.pdf")
```

```python
model = SpectralCoclustering(n_clusters = 6, random_state = 0) #clustering in 6 blocks because the wiskys come from 6 diff regions
model.fit(corr_whisky)
model.rows_
```

```python
np.sum(model.rows_, axis = 1) #shows how many whiskeys belong to each cluster
```

```python
np.sum(model.rows_, axis = 0) #shows how many clusters belong to each observation
```

```python
model.row_labels_
```

```python
whisky['Group'] = pd.Series(model.row_labels_, index = whisky.index) #extract the group labels form the model and append them to whiskey table
whisky = whisky.iloc[np.argsort(model.row_labels_)] #reorder the rows in increasing order by group labels
whisky
```

```python
whisky = whisky.reset_index(drop = True)
whisky.iloc[:, 2:14].transpose()
```

```python
correlations = pd.DataFrame.corr(whisky.iloc[:, 2:14].transpose())
correlations 
```

```python
correlations = np.array(correlations)

plt.figure(figsize=(14,7))
plt.subplot(121)
plt.pcolor(corr_whisky)
plt.title('Original')
plt.axis('tight')
plt.subplot(122)
plt.pcolor(correlations)
plt.title('Rearrange')
plt.axis('tight')
plt.show()
```
